```python

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/loan_approval_dataset - loan_approval_dataset.csv.csv")
df.columns = df.columns.str.strip()
df.head()

```

```python

sns.set_theme(style="whitegrid")
sns.countplot(x='loan_status', data=df, palette="bright")
plt.show()
```

```python
sns.set_theme(style="dark")
sns.histplot(df['cibil_score'], kde=True, color="orange")
plt.show()
```

```python
sns.boxplot(x='loan_status', y='cibil_score',
            data=df, palette="pastel")
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True),
            annot=True,
            cmap="coolwarm")
plt.show()
```

```python
import plotly.express as px
px.bar(df,
       x='loan_status',
       y='income_annum',
       color='loan_status',
       color_discrete_map={
           'Approved': 'green',
           'Rejected': 'red'
       },
       template='plotly_white')


```

```python
px.scatter(df,
           x='income_annum',
           y='loan_amount',
           color='loan_status',
           template='ggplot2')
```

```python
px.line(df,
        x='income_annum',
        y='loan_amount',
        color='loan_status',
        template='seaborn')
```

```python
px.imshow(df.corr(numeric_only=True),
          color_continuous_scale='Viridis',
          template='plotly_white')
```
